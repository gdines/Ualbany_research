### Reduce S2S Data

This will simplify the data in the regridded S2S files to look at only 1000mb temps in order to calculate a lat/lon mean and store them as new data files in atm561/final_proj/rd_data 

```python
import xarray as xr
import pandas as pd
import numpy as np
```

```python
#Reading in Datasets (odd=hightop, even=lowtop)
t_ds = xr.open_dataset("/home11/grad/2021/gd673524/research/gdines/data/realtime/ECMWF/RT_ECMWF_t.nc")
t_2ds = xr.open_dataset("/home11/grad/2021/gd673524/research/gdines/data/realtime/ECCC/RT_ECCC_t.nc")
t_3ds = xr.open_dataset("/home11/grad/2021/gd673524/research/gdines/data/realtime/UKMO/RT_UKMO_t.nc")
t_4ds = xr.open_dataset("/home11/grad/2021/gd673524/research/gdines/data/realtime/CMA/RT_CMA_t.nc")
t_5ds = xr.open_dataset("/home11/grad/2021/gd673524/research/gdines/data/realtime/JMA/RT_JMA_t.nc")
t_6ds = xr.open_dataset("/home11/grad/2021/gd673524/research/gdines/data/realtime/CNR/RT_CNR_t.nc")
```

```python
### Vertical level specificaton
vlevel = 1000
levelStr = str(vlevel)
#Changing the longitude and rerunning since the previous box didn't work
#south_boxtest = [[40,256], [40,271], [30,271], [30,256],[40,256]]
#north_box3 = [[50,252], [50,273], [40,273], [40,252],[50,252]]
#Box coordinates:
lat_min = 30
lat_mid = 40
lat_max = 50
lon_min = 252
lon_mins = 256
lon_max = 273
lon_maxs = 271
```

```python
#Reducing the Model datasets for the northern box
tn = t_ds.sel(latitude=slice(lat_max,lat_mid), longitude=slice(lon_min,lon_max), level=vlevel)#<--forecasted temps at 1000mb

t2n = t_2ds.sel(latitude=slice(lat_max,lat_mid), longitude=slice(lon_min,lon_max), level=vlevel)

t3n = t_3ds.sel(latitude=slice(lat_max,lat_mid), longitude=slice(lon_min,lon_max), level=vlevel)

t4n = t_4ds.sel(latitude=slice(lat_max,lat_mid), longitude=slice(lon_min,lon_max), level=vlevel)

t5n = t_5ds.sel(latitude=slice(lat_max,lat_mid), longitude=slice(lon_min,lon_max), level=vlevel)

t6n = t_6ds.sel(latitude=slice(lat_max,lat_mid), longitude=slice(lon_min,lon_max), level=vlevel)
```

```python
#Reducing the Model datasets for the southern box
#Reducing the Model datasets for the northern box
ts = t_ds.sel(latitude=slice(lat_mid,lat_min), longitude=slice(lon_mins,lon_maxs), level=vlevel)#<--forecasted temps at 1000mb

t2s = t_2ds.sel(latitude=slice(lat_mid,lat_min), longitude=slice(lon_mins,lon_maxs), level=vlevel)

t3s = t_3ds.sel(latitude=slice(lat_mid,lat_min), longitude=slice(lon_mins,lon_maxs), level=vlevel)

t4s = t_4ds.sel(latitude=slice(lat_mid,lat_min), longitude=slice(lon_mins,lon_maxs), level=vlevel)

t5s = t_5ds.sel(latitude=slice(lat_mid,lat_min), longitude=slice(lon_mins,lon_maxs), level=vlevel)

t6s = t_6ds.sel(latitude=slice(lat_mid,lat_min), longitude=slice(lon_mins,lon_maxs), level=vlevel)
```
