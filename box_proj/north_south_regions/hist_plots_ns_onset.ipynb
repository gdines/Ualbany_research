```python
import datetime as dt
import numpy as np
import matplotlib as mlp
mlp.use('Agg')
import matplotlib.pyplot as plt
import xarray as xr
import math
import netCDF4
import pandas as pd

%matplotlib inline
```

```python
south_per_df = pd.read_csv('/nfs/langlab_rit/gdines/box_proj/percentages/start_percentages_south.csv', 
                           index_col=False)
north_per_df = pd.read_csv('/nfs/langlab_rit/gdines/box_proj/percentages/start_percentages_north.csv', 
                           index_col=False)
```

```python
south_df = south_per_df.drop(columns=['Unnamed: 9','Unnamed: 10']).dropna()
```

```python
north_df = north_per_df.drop(columns=['Unnamed: 9','Unnamed: 10']).dropna()
```

```python
north_week_plot_data = north_df[0:4]
north_average_plot_data = north_df[4:]

south_week_plot_data = south_df[0:4]
south_average_plot_data = south_df[4:]
```

```python
#Full Event: Percentage of HT & LT Model Hits from Initialization
ax = north_week_plot_data.plot.bar(x='Time', y=['ECMWF','UKMO','JMA', 'ECCC','CMA', 'CNR', 'HT', 'LT'], rot=0,
                title='Northern Region', 
color=['orchid','purple','indigo','palegoldenrod','yellow', 'gold', 'black', 'gray']) 

plt.xlabel('Initialization in Weeks from Start of Event') 
plt.ylabel('Percentages')
fig = mlp.pyplot.gcf()
fig.set_size_inches(10, 3)
plt.ylim(0,80);
#plt.savefig("%_HT&LT_model_hits_for_initialization.jpg")
```
