### Onset of Event for Central Region

### This Code uses the percentages calculated in area_anom_b1_current.ipynb and plots them as bar graphs. 

These are for the onset of the event

```python
import datetime as dt
import numpy as np
import matplotlib as mlp
mlp.use('Agg')
import matplotlib.pyplot as plt
import xarray as xr
import math
import netCDF4
import pandas as pd

%matplotlib inline
```

```python
all_per_df = pd.read_csv('/nfs/langlab_rit/gdines/box_proj/percentages/start_percentages_central.csv', #<--file= start_percentage_central
                         names=['Time', 'ECMWF', 'ECCC', 'UKMO', 'CMA', 'JMA', 'CNR', 'HT', 'LT'], index_col=False, skiprows=1)
```

```python
week_plot_data = all_per_df[0:4]
average_plot_data = all_per_df[4:]
```

```python
#Event Onset: Percentage of HT & LT Model Hits from Initialization
ax = week_plot_data.plot.bar(x='Time', y=['ECMWF','UKMO','JMA', 'ECCC','CMA', 'CNR', 'HT', 'LT'], rot=0,
                title='Central Region', color=['orchid','purple','indigo','palegoldenrod','yellow', 'gold', 'black', 'gray']) 

plt.xlabel('Initialization in Weeks from Start of Event') 
plt.ylabel('Percentages')
fig = mlp.pyplot.gcf()
fig.set_size_inches(10, 3)
plt.ylim(0,80)
#plt.savefig("%_HT&LT_model_hits_for_initialization.jpg")
```

```python
#Start Event: Percentage of Hits per Model
ax = average_plot_data.plot.bar(x='Time', y=['ECMWF','UKMO', 'JMA', 'ECCC','CMA', 'CNR'], rot=0,
                title='Central Region', color=['orchid','purple','indigo','palegoldenrod','yellow', 'gold'])
plt.xlabel('') 
plt.ylabel('Percentages')
plt.ylim(0,35)
fig = plt.gcf()
fig.set_size_inches(5, 4, forward=True);
#plt.savefig("%_HT&LT_model_hits.jpg")
```

```python
#Start Event: Percentage of HT vs. LT Model Hits
ax = average_plot_data.plot.bar(x='Time', y=['HT', 'LT'], rot=0, title='Central Region', 
                  color=['purple','gold']) 
plt.xlabel('') 
plt.ylabel('Percentages')
plt.ylim(0,24)
fig = plt.gcf()
fig.set_size_inches(5, 4, forward=True);
```
